# Master dictonary

```python
# Especially in sentiment analysis "Master Dictionary - is used to identify whether the words are positive or negative
# Positive Words: Words that are associated with positive sentiments or emotions. Examples include "happy", "joyful", "excited", etc.
# Negative Words: Words that are associated with negative sentiments or emotions. Examples include "sad", "angry", "frustrated", etc.
```

```python
import os

# Define the folder path
folder_path = r"C:\Users\harik\OneDrive\Pictures\Downloads\MasterDictionary-20240818T081146Z-001\MasterDictionary"

# Initialize sets to store positive and negative words
positive_words = set()
negative_words = set()

# Iterate over all files in the folder
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):  # Check if it's a file
        with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
            # Read each line and process it
            for line in file:
                parts = line.strip().split('\t')  # Adjust split delimiter as needed
                if len(parts) == 2:
                    word, sentiment = parts
                    if sentiment == 'positive':
                        positive_words.add(word)
                    elif sentiment == 'negative':
                        negative_words.add(word)
                else:
                    print(f"{line.strip()}")

# Now, `positive_words` and `negative_words` contain unique words from the Master Dictionary
print(f"Total positive words loaded: {len(positive_words)}")
print(f"Total negative words loaded: {len(negative_words)}")

# Print the actual positive words
print("\nPositive Words:")
for word in sorted(positive_words):
    print(word)

# Print the actual negative words
print("\nNegative Words:")
for word in sorted(negative_words):
    print(word)
```

```python
import os
import pandas as pd
from textblob import TextBlob
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import cmudict
from nltk import download, pos_tag
from openpyxl import load_workbook
from openpyxl.utils.dataframe import dataframe_to_rows

# Download necessary resources
download('punkt')
download('cmudict')
download('averaged_perceptron_tagger')

# Define paths
folder_path = r"C:\Users\harik\OneDrive\Pictures\Downloads\Extracted_Data"
stopwords_folder_path = r"C:\Users\harik\OneDrive\Pictures\Downloads\StopWords-20240820T085234Z-001\StopWords"
positive_words_file = r"C:\Users\harik\OneDrive\Pictures\Downloads\MasterDictionary-20240818T081146Z-001\MasterDictionary\positive-words.txt"
negative_words_file = r"C:\Users\harik\OneDrive\Pictures\Downloads\MasterDictionary-20240818T081146Z-001\MasterDictionary\negative-words.txt"
```

```python
# Load Stop Words
stop_words = set()
for filename in os.listdir(stopwords_folder_path):
    with open(os.path.join(stopwords_folder_path, filename), 'r', encoding='utf-8', errors='ignore') as file:
        stop_words.update(file.read().splitlines())

# Load Master Dictionary
positive_words = set()
negative_words = set()

# Load positive words
with open(positive_words_file, 'r', encoding='utf-8', errors='ignore') as file:
    positive_words.update(file.read().splitlines())

# Load negative words
with open(negative_words_file, 'r', encoding='utf-8', errors='ignore') as file:
    negative_words.update(file.read().splitlines())
```

```python
# Load CMU dictionary for syllable count
d = cmudict.dict()

def syllable_count(word):
    word = word.lower()
    if word in d:
        return [len([y for y in x if y[-1].isdigit()]) for x in d[word]][0]
    return 0

def readability_metrics(text):
    words = word_tokenize(text)
    sentences = sent_tokenize(text)
    words = [word for word in words if word.isalnum() and word.lower() not in stop_words]
    
    # Calculate sentiment metrics
    positive_score = sum(1 for word in words if word.lower() in positive_words)
    negative_score = sum(1 for word in words if word.lower() in negative_words)
    polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)
    subjectivity_score = (positive_score + negative_score) / (len(words) + 0.000001)
    
    # Readability metrics
    word_count = len(words)
    sentence_count = len(sentences)
    avg_sentence_length = word_count / sentence_count if sentence_count > 0 else 0
    avg_words_per_sentence = word_count / sentence_count if sentence_count > 0 else 0
    complex_word_count = sum(1 for word in words if syllable_count(word) > 2)
    syllables_per_word = sum(syllable_count(word) for word in words) / word_count if word_count > 0 else 0
    fog_index = 0.4 * (avg_sentence_length + (100 * complex_word_count / word_count))
    complex_words_percentage = (complex_word_count / word_count) * 100
    
    # Calculate personal pronouns
    pos_tags = pos_tag(words)
    personal_pronouns = sum(1 for word, tag in pos_tags if tag in ['PRP', 'PRP$', 'WP', 'WP$'])
    
    # Calculate average word length
    avg_word_length = sum(len(word) for word in words) / word_count if word_count > 0 else 0
    
    return {
        'URL_ID': '',  # Placeholder for URL_ID
        'POSITIVE SCORE': positive_score,
        'NEGATIVE SCORE': negative_score,
        'POLARITY SCORE': polarity_score,
        'SUBJECTIVITY SCORE': subjectivity_score,
        'AVG SENTENCE LENGTH': avg_sentence_length,
        'AVG WORDS PER SENTENCE': avg_words_per_sentence,
        'PERCENTAGE OF COMPLEX WORDS': complex_words_percentage,
        'FOG INDEX': fog_index,
        'COMPLEX WORD COUNT': complex_word_count,
        'WORD COUNT': word_count,
        'SYLLABLE PER WORD': syllables_per_word,
        'PERSONAL PRONOUNS': personal_pronouns,
        'AVG WORD LENGTH': avg_word_length
    }

# Process each text file and compute metrics
data = []
for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        file_id = os.path.splitext(filename)[0]
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
            text = file.read()
            metrics = readability_metrics(text)
            metrics['URL_ID'] = file_id
            data.append(metrics)
```

```python
# Create DataFrame
df = pd.DataFrame(data)

# Save to Excel
output_path = r"C:\Users\harik\OneDrive\Pictures\Downloads\Text_Analysis_Results.xlsx"
df.to_excel(output_path, index=False)

# Adjust column widths
wb = load_workbook(output_path)
ws = wb.active

# Calculate max length of each column
for col in ws.columns:
    max_length = 0
    column = col[0].column_letter  # Get column name
    for cell in col:
        try:
            if len(str(cell.value)) > max_length:
                max_length = len(cell.value)
        except:
            pass
    adjusted_width = (max_length + 2)  # Add a little extra space
    ws.column_dimensions[column].width = adjusted_width

# Save adjusted Excel file
wb.save(output_path)

print("Text Analysis completed and results saved to Excel with adjusted column widths.")
```
